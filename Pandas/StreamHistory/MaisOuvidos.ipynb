```python
import pandas as pd
import numpy as np
```

```python
stream0 = spark.read.option("multiline", "true").option("mode", "PERMISSIVE").json("dbfs:/FileStore/tables/SpotifyData/StreamingHistory0.json").toPandas()
stream1 = spark.read.option("multiline", "true").option("mode", "PERMISSIVE").json("dbfs:/FileStore/tables/SpotifyData/StreamingHistory1.json").toPandas()
stream2 = spark.read.option("multiline", "true").option("mode", "PERMISSIVE").json("dbfs:/FileStore/tables/SpotifyData/StreamingHistory2.json").toPandas()
stream3 = spark.read.option("multiline", "true").option("mode", "PERMISSIVE").json("dbfs:/FileStore/tables/SpotifyData/StreamingHistory3.json").toPandas()
```

```python
frames = [stream0, stream1, stream2, stream3]
streamHistory = pd.concat(frames)
streamHistory.head()
```

```python
streamHistory.shape
```

```python
streamHistory.columns
```

```python
streamHistory.columns = ['artistName', 'date', 'minutesPlayed', 'trackName']
```

```python
streamHistory['minutesPlayed'] = streamHistory['minutesPlayed']/60000
streamHistory['minutesPlayed'] = streamHistory['minutesPlayed'].round(decimals=2)
streamHistory
```

```python
streamHistory.dtypes
```

```python
streamHistory['date'] = pd.to_datetime(streamHistory.date)
streamHistory.head()
```

```python
streamHistory.dtypes
```

```python
streamHistory['year'] = streamHistory.date.dt.year
streamHistory['year'] = streamHistory['year'].astype('str')
streamHistory.head()
```

```python
streamHistory.describe()
```

```python
streamHistory = streamHistory[(streamHistory.trackName != 'Unknown Track') | (streamHistory.artistName != 'Unknown Artist')]
streamHistory = streamHistory[streamHistory.minutesPlayed != 0]
```

```python
tracks = streamHistory.groupby(['trackName', 'artistName', 'year'], as_index=False).sum()
tracks = tracks.sort_values(['minutesPlayed'], ascending=False).reset_index(drop=True)
tracks.index = tracks.index + 1
tracks.head()
```

```python
tracks2020Top10 = tracks[tracks.year == '2020']
tracks2020Top10 = tracks2020Top10.iloc[0:10].reset_index(drop=True)
tracks2020Top10.index = tracks2020Top10.index + 1
tracks2020Top10
```

```python
tracks2021Top10 = tracks[tracks.year == '2021']
tracks2021Top10 = tracks2021Top10.iloc[0:10].reset_index(drop=True)
tracks2021Top10.index = tracks2021Top10.index + 1
tracks2021Top10
```

```python
artists = streamHistory.groupby(['artistName', 'year'], as_index=False).sum()
artists = artists.sort_values(['minutesPlayed'], ascending=False).reset_index(drop=True)
artists.index = artists.index + 1
artists.head()
```

```python
artists2020Top10 = artists[artists.year == '2020']
artists2020Top10 = artists2020Top10.iloc[0:10].reset_index(drop=True)
artists2020Top10.index = artists2020Top10.index + 1
artists2020Top10
```

```python
artists2021Top10 = artists[artists.year == '2021']
artists2021Top10 = artists2021Top10.iloc[0:10].reset_index(drop=True)
artists2021Top10.index = artists2021Top10.index + 1
artists2021Top10
```
